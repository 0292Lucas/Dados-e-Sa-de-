 O pvalor é muito baixo, então podemos supor que as variâncias não são as mesmas. Neste caso, vamos usar o Welch ANOVA.

```python
pg.welch_anova(dv='pontos_PHQ', between='ADHERENCE', data=df_agrupamento)
```

Como o pvalor é muito baixo, podemos rejeitar a hipótese nula e concluir que todos os homens não são iguais. Agora, vamos ver quais são as diferenças. Vamos usar o teste Games Howell para ver as diferenças. A razão pela qual escolhemos este teste é porque ele não assume normalidade, mesma variância e tamanho das amostras como iguais.

```python
pg.pairwise_gameshowell(dv='pontos_PHQ', between='ADHERENCE', data=df_agrupamento)
```

Pvalor nos grupos 1 e 2 é muito baixo, então podemos assumir que as médias são diferentes.

Pvalor nos grupos 1 e 3 também é baixo, então podemos assumir que suas médias também são diferentes.

Mas o valor P nos grupos 2 e 3 não é baixo, então nesses grupos podemos assumir que as médias são iguais.

Com isso, podemos dizer que praticar pelo menos 150 minutos de exercício por semana tem um impacto benéfico no escore de php.
